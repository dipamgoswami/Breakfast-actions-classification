```
import csv
pred = []
gt = []
with open('results/predictions_final-segment-lstm_20210220_143509_0.575.h520210222_102403.csv','r') as f2:
    readfile = csv.reader(f2)
    for line in readfile:
        if line[1]!='Category':
            pred.append(int(line[1]))

with open('data/test_segment_labels.csv','r') as f1:
    readfile = csv.reader(f1)
    for line in readfile:
        gt.append(int(line[1]))

from sklearn.metrics import accuracy_score
accuracy_score(gt,pred)
```
